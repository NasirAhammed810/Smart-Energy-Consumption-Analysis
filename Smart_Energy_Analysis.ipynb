```python
#Feature Engineering
daily_df = df.resample('D').sum()

```

```python
features = daily_df.copy()
```

```python
# Day of month (1â€“31)
features['day'] = features.index.day

# Month (1â€“12)
features['month'] = features.index.month

# Day of week (0 = Monday, 6 = Sunday)
features['day_of_week'] = features.index.dayofweek

# Weekend flag (Saturday & Sunday)
features['is_weekend'] = features['day_of_week'].apply(
    lambda x: 1 if x >= 5 else 0
)
```

```python
# Previous day energy consumption
features['lag_1'] = features['total_energy_consumption'].shift(1)

# Energy consumption 7 days ago
features['lag_7'] = features['total_energy_consumption'].shift(7)

```

```python
# 7-day rolling mean (weekly trend)
features['rolling_mean_7'] = features['total_energy_consumption'].rolling(7).mean()

# 14-day rolling mean (bi-weekly trend)
features['rolling_mean_14'] = features['total_energy_consumption'].rolling(14).mean()

# 7-day rolling standard deviation (volatility)
features['rolling_std_7'] = features['total_energy_consumption'].rolling(7).std()

```

```python
features = features.dropna()
```

```python
features.head()
features.columns
features.shape

```

```python
# Temperature change
features['temp_diff'] = features['temperature'].diff()

# Humidity change
features['humidity_diff'] = features['humidity'].diff()
```

```python
features.dropna(inplace=True)
```

```python
#Aggregation


# List of device-level energy columns
device_columns = [
    'fridge',
    'wine_cellar',
    'garage_door',
    'microwave',
    'living_room_appliances'
]




```

```python
device_long_df = daily_df[device_columns].reset_index().melt(
    id_vars='timestamp',
    value_vars=device_columns,
    var_name='device',
    value_name='energy_kwh'
)
```

```python
device_energy_mean = (
    device_long_df
    .groupby('device')['energy_kwh']
    .mean()
    .reset_index()
)
device_energy_mean
```

```python
#module 4

# Target variable (what we want to predict)
y = features['total_energy_consumption']

# Input features (all except target)
X = features.drop(columns=['total_energy_consumption'])

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    shuffle=False
)
```

```python
#Implementing Linear regression


from sklearn.linear_model import LinearRegression

# Initialize model
lr_model = LinearRegression()

# Train model
lr_model.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(y_test.values, label="Actual Energy Consumption")
plt.plot(y_pred, label="Predicted Energy Consumption")
plt.title("Baseline Linear Regression â€“ Daily Energy Prediction")
plt.xlabel("Time")
plt.ylabel("Energy Consumption (kWh)")
plt.legend()
plt.show()
```

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lr_model.coef_
})

feature_importance.sort_values(by='Coefficient', ascending=False)
```

```python
results = X_test.copy()
results['Actual_Energy'] = y_test.values
results['Predicted_Energy'] = y_pred

results.head()
```
